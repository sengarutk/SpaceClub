```python
#printing graph
import matplotlib.pyplot as plt
from astropy.io import fits

fits_file = 'bonus_image_1.fits'

with fits.open(fits_file, mode = 'readonly') as hdul: 
   eta = hdul[1].data['ETA']
   eta_corr = hdul[1].data['ETA_CORR']

# Plotting the graph
plt.plot(eta, eta_corr, marker='o', linestyle='-', color='black')
plt.title('ETA vs ETA_CORR')
plt.xlabel('ETA')
plt.ylabel('ETA_CORR')
plt.grid(True)
plt.show()
```
