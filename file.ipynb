```python
from astropy.io import fits
f = fits.open('bonus_image_1.fits') 
f.info() 
```

```python
imdata = f[0].data
print(imdata) 
cols = f[1].columns
print(cols)
cols.info()
```

```python
import matplotlib.pyplot as plt 
from matplotlib.colors import LogNorm
plt.figure()
plt.imshow(imdata,cmap = 'ocean', origin = 'lower', norm = LogNorm())
plt.colorbar()
plt.show()
```

```python
from astropy.io import fits 
f1 = fits.open('lightcurve_1.fits')
f1.info()
```

```python
cols = f1[1].columns
print(cols)
```

```python
time = f1[1].data['TIME']
sap1 = f1[1].data['PDCSAP_FLUX']

plt.xlabel('TIME')
plt.ylabel('PDCSAP_FLUX')
plt.title('PDCSAP_FLUX VS TIME') 

plt.plot(time, sap1, 'ko')
plt.show()
```

```python
import numpy as np 
time = f1[1].data['TIME']
sap1 = f1[1].data['PDCSAP_FLUX']
sap2 = f1[1].data['SAP_QUALITY']

plt.xlabel('TIME')
plt.ylabel('PDCSAP_FLUX')
plt.title('PDCSAP_FLUX VS TIME') 
plt.legend()

plt.plot(time, sap1, 'ko')

wp = np.where(sap2 > 0)[0]
plt.plot(time[wp], sap1[wp], 'ro')
plt.grid(True)
plt.show()
```
